# 8 - Split the data into dependent and independent variables

```python
df=df.drop(['Age'], axis=1)
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

# 9 - Scale the independent variables

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = scale.fit_transform(X)
X
```

# 10 - Splitting the test and train variables

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y ,random_state=1, shuffle=True, train_size = 0.75)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# 11 and 12 - Build and train the model

```python
from sklearn import linear_model
reg = linear_model.LogisticRegression()
reg.fit(X_train, y_train)
```

# 13 and 14 - Test the model and measure performance using metrics

```python
y_pred = reg.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
